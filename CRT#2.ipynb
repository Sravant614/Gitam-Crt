#Contact Application
- Add contact 
- Search the contact list
- list all the contacts(merge/import)
- name 1:value
- name 1:value
- Modify,delete,update the contact

```
contacts={}
def AddContact(name,phone):
    if name not in contacts:
        contacts[name]=phone
        print("contacts are added")
    else:
        print("not added")
    return
AddContact("XYZ",'4375847545')
```

```
contacts
```

```
def searchContact(name):
    if name in contacts:
        print(name,":",contacts[name])
    else:
        print("not found")
    return
searchContact('XYZ')
```

```
def importContact(newContacts):
    contacts.update(newContacts)
    print(len(newContacts.keys()),'contacts are added')
    return
newContacts={"sai":55677897987}
importContact(newContacts)
```

```
def deleteContact(name):
    if name in contacts:
        del contacts[name]
    else:
        print("not found")
name=input("Enter name to delete:")
deleteContact(name)
print(contacts)    
```
